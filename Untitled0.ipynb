```python

import pandas as pd

# Read the uploaded CSV
df = pd.read_csv("products.csv", parse_dates=["Date"])

# Show first few rows
df.head()
```

```python

# Check info and summary
print(df.info())
print(df.describe())

# Ensure Price is numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Drop rows with missing prices
df = df.dropna(subset=['Price'])
```

```python

# Find the store with the lowest price for each product
idx = df.groupby("Product")["Price"].idxmin()
best_prices = df.loc[idx].reset_index(drop=True)

best_prices
```

```python

import matplotlib.pyplot as plt

# Bar chart for each product
for product in df['Product'].unique():
    subset = df[df['Product'] == product]
    plt.figure()
    plt.bar(subset['Store'], subset['Price'])
    plt.title(f'Prices for {product}')
    plt.ylabel('Price')
    plt.xlabel('Store')
    plt.show()

# Box plot of price distribution
plt.figure(figsize=(8,4))
df.boxplot(column='Price', by='Product', grid=False)
plt.title('Price distribution by product')
plt.suptitle('')
plt.show()
```

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Average price per product
avg_price = df.groupby('Product')['Price'].mean().reset_index()
X = avg_price[['Price']].values

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans (3 groups: low, mid, high)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

avg_price['Cluster'] = labels
avg_price
```

```python

plt.scatter(avg_price['Product'], avg_price['Price'],
            c=avg_price['Cluster'], cmap='viridis', s=100)
plt.xticks(rotation=45)
plt.title("Product Clusters by Price")
plt.ylabel("Average Price")
plt.show()
```

```python

# 1. Install mlxtend (only once per session)
!pip install mlxtend

# 2. Imports
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# 3. Load transaction CSV
tdata = pd.read_csv("transactions.csv")

# 4. Convert products into list of lists
transactions = [row.split(";") for row in tdata['Products']]

# 5. Encode transactions
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
trans_df = pd.DataFrame(te_ary, columns=te.columns_)

# 6. Frequent itemsets (lower support to 0.05 if needed)
freq_itemsets = apriori(trans_df, min_support=0.05, use_colnames=True)

# 7. Association rules
if not freq_itemsets.empty:
    rules = association_rules(freq_itemsets, metric="confidence", min_threshold=0.3)
    print(rules[['antecedents','consequents','support','confidence','lift']].head())
else:
    print("⚠️ No frequent itemsets found. Try lowering min_support.")
```
